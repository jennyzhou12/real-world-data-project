```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df16=pd.read_csv("en_climate_daily_ON_6111769_2016_P1D.csv")
df17=pd.read_csv("en_climate_daily_ON_6111769_2017_P1D.csv")
df18=pd.read_csv("en_climate_daily_ON_6111769_2018_P1D.csv")
df19=pd.read_csv("en_climate_daily_ON_6111769_2019_P1D.csv")
df20=pd.read_csv("en_climate_daily_ON_6111769_2020_P1D.csv")
df16.columns
```

```python
df16end=df16[(df16['Date/Time'] >= '2016-12-01') & (df16['Date/Time'] <='2016-12-31')]
df17end=df17[(df17['Date/Time'] >= '2017-12-01') & (df17['Date/Time'] <='2017-12-31')]
df18end=df18[(df18['Date/Time'] >= '2018-12-01') & (df18['Date/Time'] <='2018-12-31')]
df19end=df19[(df19['Date/Time'] >= '2019-12-01') & (df19['Date/Time'] <='2019-12-31')]
```

```python
df17begin=df17[(df17['Date/Time'] >= '2017-01-01') & (df17['Date/Time'] <='2017-03-31')]
df18begin=df18[(df18['Date/Time'] >= '2018-01-01') & (df18['Date/Time'] <='2018-03-31')]
df19begin=df19[(df19['Date/Time'] >= '2019-01-01') & (df19['Date/Time'] <='2019-03-31')]
df20begin=df20[(df20['Date/Time'] >= '2020-01-01') & (df20['Date/Time'] <='2020-03-31')]
```

```python
# yr1617=[df16end,df17begin]
# yr1718=[df17end,df18begin]
# yr1819=[df18end,df19begin]
# yr1920=[df19end,df20begin]
result = pd.concat([df16end,df17begin,df17end,df18begin,df18end,df19begin,df19end,df20begin])
dataneed=result["Min Temp (°C)"]
```

```python
dataneed = dataneed.reset_index(drop=True)
```

```python
dataneed.plot(figsize=(70,10))
plt.axhline(y=0, color='r', linestyle='-')
plt.show()
```
